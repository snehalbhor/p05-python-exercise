```python
#question no, 17 *********************************************************
dic={'aks':70, 'ram':20, 'sah':30, 'mahesh':40}
print(dic)
def returnsum(one):
    list=[]
    for i in one:
        list.append(one[i])
    final=sum(list)

    return final

res=returnsum(dic)
print(res)


def returnSum(myDict):
 
    list = []
    for i in myDict:
        list.append(myDict[i])
    final = sum(list)
 
    return final
 
# Driver Function
dict = {'a': 100, 'b': 200, 'c': 300}
print("Sum :", returnSum(dict))
```
